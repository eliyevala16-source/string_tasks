```python
def yoxla(name):
    no = '\\/:*?"<>|'
    if len(name) > 10:
        print("Ad 10 simvoldan artıqdır")
        return
    for x in name:
        if x in no:
            print("Yanlış simvol var")
            return
    print("Düzgün fayl adı")
name = input("Faylın adını daxil edin: ")
yoxla(name)
```

```python
def cem(ifade):
    total = 0
    reqem = ""
    for x in ifade:
        if x != "+":
            reqem += x
        else:
            total += int(reqem)
            reqem = ""
    total += int(reqem)
    return total
ifade = input("İfadeni daxil edin: ")
print("Cavab:", cem(ifade))
```

```python
def birinci_soz(text):
    soz = ""
    for x in text:
        if x != " ":
            soz += x
        else:
            break
    return soz
setir = input("Sətri daxil edin: ")
print("Birinci söz:", birinci_soz(setir))
```

```python
def qovluq(name):
    special = ".-~_=:"
    if len(name) < 5 or len(name) > 15:
        print("5<=Simvol sayı<=15 olmalıdır!")
        return
    boyuk = False
    kicik = False
    reqem = False
    xususi = False
    for x in name:
        if "A" <= x <= "Z":
            boyuk = True
        elif "a" <= x <= "z":
            kicik = True
        elif "0" <= x <= "9":
            reqem = True
        elif x in special:
            xususi = True
    if not reqem:
        print("Yanlış! Ən az 1 rəqəm olmalıdır!")

    elif not xususi:
        print("Yanlış! Ən az 1 xususi simvol olmalıdır!")

    elif not boyuk or not kicik:
        print("Ən az 1 böyük və 1 kiçik hərf olmalıdır!")

    else:
        print("Düzgün qovluq adı!")


name = input("Qovluq adi daxil edin: ")
qovluq(name)
```
